# Проект по спринту "Основы машинного обучения"

**Описание проекта:**

Сеть фитнес-центров «Культурист-датасаентист» разрабатывает стратегию взаимодействия с клиентами на основе аналитических данных.
Распространённая проблема фитнес-клубов и других сервисов — отток клиентов. Как понять, что клиент больше не с вами? Можно записать в отток тех, кто попросил закрыть договор или удалил аккаунт. Однако клиенты не всегда уходят демонстративно: чаще перестают пользоваться сервисом тихо.

Для фитнес-центра можно считать, что клиент попал в отток, если за последний месяц ни разу не посетил спортзал. Конечно, не исключено, что он уехал на Бали и по приезде обязательно продолжит ходить на фитнес. Однако чаще бывает наоборот. Если клиент начал новую жизнь с понедельника, немного походил в спортзал, а потом пропал — скорее всего, он не вернётся.

Чтобы бороться с оттоком, отдел по работе с клиентами «Культуриста-датасаентиста» перевёл в электронный вид множество клиентских анкет. 

**Задача**: провести анализ и подготовить план действий по удержанию клиентов. А именно:
* cпрогнозировать вероятность оттока (на уровне следующего месяца) для каждого клиента;
* сформировать типичные портреты клиентов: выделить несколько наиболее ярких групп и охарактеризовать их основные свойства;
* проанализировать основные признаки, наиболее сильно влияющие на отток;
* сформулировать основные выводы и разработать рекомендации по повышению качества работы с клиентами (выделить целевые группы клиентов, предложить меры по снижению оттока, определить другие особенности взаимодействия с клиентами).

* `age` — возраст;
* `lifetime` — время с момента первого обращения в фитнес-центр (в месяцах).

Информация на основе журнала посещений, покупок и информация о текущем статусе абонемента клиента:
* `contract_period` — длительность текущего действующего абонемента (месяц, 6 месяцев, год);
* `month_to_end_contract` — срок до окончания текущего действующего абонемента (в месяцах);
* `group_visits` — факт посещения групповых занятий;
* `avg_class_frequency_total` — средняя частота посещений в неделю за все время с начала действия абонемента;
* `avg_class_frequency_current_month` — средняя частота посещений в неделю за предыдущий месяц;
* `avg_additional_charges_total` — суммарная выручка от других услуг фитнес-центра: кафе, спорттовары, косметический и массажный салон.


* `churn` — факт оттока в текущем месяце.

### Матрица корреляций

Построим матрицу корреляций с помощью метода `corr()` и по ней тепловую карту `heatmap()` библиотеки `seaborn`. Создадим для этого функцию `display_corr()`.

```python
def display_corr(data, title):
    corr_data = data.corr()
    display(corr_data)
    plt.figure(figsize = (12,12))
    sns.heatmap(corr_data, square = True, annot = True)
    plt.title('Тепловая карта корреляции', size=16)
    plt.show()
    
    
display_corr(data, 'Тепловая карта корреляции')
```

Судя по матрице корреляции есть пара признаков у которых корреляция высокая 0,95-0,97. Эти признаки мультиколлинеарны. 

Мультиколлинеарные пары:
- `avg_class_frequency_total` и `avg_class_frequency_current_month`;
- `month_to_end_contract` и `contract_period`.

Оставим из мультиколлинеарных пар значения средней частоты посещений в неделю за предыдущий месяц `avg_class_frequency_current_month` и длительность текущего действующего абонемента (месяц, 6 месяцев, год) `contract_period`. Уберем лишние признаки для дальнейшей работы с помощью метода `drop()`, сохраним изменения в переменную `new_data`. Построим тепловую карту еще раз, используя фунцию `display_corr()`.

```python
new_data = data.drop(columns=['avg_class_frequency_total', 'month_to_end_contract'])
display_corr(new_data, 'Тепловая карта корреляции')
```

У некоторых признаков есть небольшая корреляция, модуль больше 0,4. Этого недостаточно для того, чтобы говорить о прямой корреляции, но может есть зависимость этих признаков.

Модуль корреляции больше 40% у следующих признаков:
- `partner` и `promo_friends` - прямая корреляция;

## Построение модели прогнозирования оттока клиентов

Целевой признак исследования - факт оттока клиента в следующем месяце, переменная `churn`. Это задача бинарной классификации клиентов.

Сохраним признаки в переменной `X`, а целевую величину — в переменной `y`. Разобьем данные на обучающую и валидационную выборку функцией `train_test_split()` в соотношении 80:20.

```python
X = new_data.drop('churn', axis=1)
y = new_data['churn']

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify = y, random_state = 0, test_size = 0.2)
```

Обучим модель на train-выборке двумя способами:
- логистической регрессией,
- случайным лесом.

Логистическая регрессия популярный метод решения задач при бинарной классификации. Она реализована как класс `LogisticRegression()` в модуле `linear_model` библиотеки `sklearn`.

Случайный лес генерирует множество различных независимых друг от друга деревьев слегка разными способами (берёт разные подвыборки, разные признаки), а на основании их ответов формирует итоговое решение. Алгоритм случайного леса выбирает голосованием (в классификации) тот ответ, который большинство деревьев в лесу считает правильным. Случайный лес в `sklearn` реализован в модуле `ensemble`, для задачи классификации применяют алгоритм `RandomForestClassifier()`.

Для сравнения качества моделей рассмотрим метрики `accuracy, precision и recall` для обеих моделей на валидационной выборке.

Создадим функцию `make_prediction()` для обучения модели, создания прогноза и вычисления метрик на валидационной выборке. Передадим в функцию модели логической регрессии и случайного леса.

```python
# функция, которая принимает на вход модель и данные и выводит метрики классификации
def make_prediction(m, X_train, X_test, y_train, y_test):
    model = m
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print('Модель', m)
    print('Accuracy: {:.3f}'.format(accuracy_score(y_test, y_pred)))
    print('Precision: {:.3f}'.format(precision_score(y_test, y_pred)))
    print('Recall: {:.3f}\n'.format(recall_score(y_test, y_pred)))
    

# модель логистичческой регрессии
make_prediction(LogisticRegression(solver='liblinear', random_state = 0), 
                X_train, X_test, y_train, y_test)

# модель слуачйного леса
make_prediction(RandomForestClassifier(n_estimators = 100, random_state = 0), 
                X_train, X_test, y_train, y_test)
```

Доля правильных ответов `Accuracy`, точность `Precision` и полнота `Recall` немного лучше у модели случайного леса. Логично, что ансамбли дают более точный результат, хотя линейные модели тоже неплохи.

Для решения задачи бинарной классификации клиентов использованы модели логистической регрессии `LogisticRegression()` и случайного леса `RandomForestClassifier()`.

Для сравнения точносчти моделей расчитаны метрики `accuracy, precision и recall` для обеих моделей на валидационной выборке. По результатам обе модели достаточно точные, но лучше модель случайного леса `RandomForestClassifier()`:

## Кластеризация клиентов

### Выделение числа кластеров

Кластеризация — разделение объектов на группы. Главное отличие кластеризации от классификации — отсутствие заранее заданных классов, или «правильных ответов». Кластеризируем клиентов фитнес-центра.

Удалим из данных столбец `churn` и сохраним в переменную `cluster_data`. Для начала стандартизируем данные. В `sklearn` для этого в модуле `preprocessing` есть готовый класс `StandardScaler()`. Они напоминают модели: их также нужно обучать — показывать, какие значения принимает признак на примере обучающей выборки, — и только после этого применять к любым новым наблюдениям.

```python
# создание final_data
cluster_data = new_data.drop(columns = ['churn'])
# стандартизация данных
sc = StandardScaler()
x_sc = sc.fit_transform(cluster_data)
```

Иерархическая агломеративная кластеризация интуитивно понятная и наглядная. На каждом шаге алгоритм наращивает кластеры, присоединяя соседние. Потому он и называется агломеративным, ведь превращает кластеры в агломерации. Важно понять — в какой именно момент процесс объединения следует остановить. 

По дендрограмме можно визуально оценить, сколько кластеров должно быть. А также прикинуть расстояние, после которого мы перестаём объединять объекты.

Нарисуем дендограмму с помощью `linkage()` и `dendrogram()` из библиотеки `scipy.cluster.hierarchy`.

В переменной `linked` сохраним таблица «связок» между объектами. Её можно визуализировать как дендрограмму.

```python
linked = linkage(x_sc, method = 'ward')
plt.figure(figsize=(15, 10))  
dendrogram(linked, orientation='top')
plt.title('Иерархическая агломеративная кластеризация фитнес-центра')
plt.show()
```

Предложенное оптимальное число кластеров 4 — четыре разных цветов на графике. Выберем знчение 5, количество линий пересекаемых при значении 60, так как фиолетовый класстер достаточно большой.

### Кластеризация алгоритмом K-Means 

K-Means группирует объекты пошагово. Алгоритм основан на предположении, что число кластеров (групп) заранее известно. В предыдущем пункте с помощью иерархической агломеративной кластеризации, нами получено количество кластеров - 5.

Принцип работы пошагового алгоритма K-means:
1. Есть K кластеров. Алгоритм пошагово подбирает их центры и относит объекты к тому кластеру, чей центр ближе.
2. Центры корректируются (меняют своё положение) до тех пор, пока это позволяет минимизировать среднее расстояние от объектов каждого кластера до его центра.
3. Когда расстояние от объектов до центра перестаёт снижаться или сокращается несущественно, алгоритм останавливается и фиксирует разбиение, считая его оптимальным.

Алгоритм под названием `KMeans` реализован в модуле `sklearn.cluster`. Cохраним метки кластера в поле датасета `cluster`.

```python
# задаём число кластеров, равное 5, и фиксируем значение random_state
km = KMeans(n_clusters=5, random_state=0) 
# применяем алгоритм к данным и формируем вектор кластеров 
labels = km.fit_predict(x_sc)
# сохраняем метки кластера в поле датасета
new_data['cluster'] = labels
new_data.head()
```
